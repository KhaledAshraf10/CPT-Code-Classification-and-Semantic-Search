#Preprocessing and Cleaning

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    # Add space after numbers if not present
    text = re.sub(r'(\d+)(?=\D)', r'\1 ', text)
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)

# Apply cleaning
df_notes['Description'] = df_notes['Description'].apply(clean_text)
df_formal['Cleaned_Description'] = df_formal['Formal Description'].apply(clean_text)
```

```python

def plot_training_curves(history):
    """
    Plot training and validation metrics from history object.

    Args:
        history: History object returned from model.fit() or a dictionary
                containing 'loss', 'val_loss', etc.
    """
    # Extract metrics
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    # Accuracy if available
    acc = history.history.get('accuracy', None)
    val_acc = history.history.get('val_accuracy', None)

    epochs = range(1, len(loss) + 1)

    # Create figure
    plt.figure(figsize=(12, 5))

    # Plot loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, loss, 'bo-', label='Training loss')
    plt.plot(epochs, val_loss, 'ro-', label='Validation loss')
    plt.title('Training and validation loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Plot accuracy if available
    if acc is not None:
        plt.subplot(1, 2, 2)
        plt.plot(epochs, acc, 'bo-', label='Training acc')
        plt.plot(epochs, val_acc, 'ro-', label='Validation acc')
        plt.title('Training and validation accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
# Function to plot learning curves
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")

    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1,
                     color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

 # FastText Embedding and Model

```python
from gensim.models import FastText

# Tokenize sentences first
sentences = df_notes['Description'].apply(lambda x: x.split()).tolist()

# Train FastText model
fasttext_model = FastText(sentences, vector_size=100, window=3, min_count=1, workers=4)

# Create sentence embeddings (average word vectors)
def get_fasttext_sentence_vector(text):
    tokens = text.split()
    vectors = []
    for token in tokens:
        if token in fasttext_model.wv:
            vectors.append(fasttext_model.wv[token])
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(100)

X_embed_fasttext = np.vstack(df_notes['Description'].apply(get_fasttext_sentence_vector))

# Split
X_train, X_test, y_train, y_test = train_test_split(X_embed_fasttext, df_notes['Code'], test_size=0.2, random_state=42, stratify=df_notes['Code'])

# Train Logistic Regression
model_ft = LogisticRegression(max_iter=1000)
model_ft.fit(X_train, y_train)

# Predict
y_pred_ft = model_ft.predict(X_test)

# Evaluate
print("✅ FastText Logistic Regression Results:")
print(classification_report(y_test, y_pred_ft))
sns.heatmap(confusion_matrix(y_test, y_pred_ft), annot=True, fmt='d')
plt.title('Confusion Matrix - FastText Embedding')
plt.show()

```

# Word2Vec Embedding and Machine learning Models

```python
from gensim.models import Word2Vec
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

#  Train Word2Vec model
sentences = [note.split() for note in df_notes['Description']]
w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

#  Create sentence embeddings
def get_w2v_sentence_vector(text):
    tokens = text.split()
    vectors = [w2v_model.wv[token] for token in tokens if token in w2v_model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(100)

X_embed_w2v = np.vstack(df_notes['Description'].apply(get_w2v_sentence_vector))

#  Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(df_notes['Code'])

#  Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_embed_w2v, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

#  Classifiers to compare
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "SVM": SVC(kernel='linear', probability=True)
}

# training and Evaluation
for model_name, model in models.items():
    print(f"\n🔍 Training {model_name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Decode predictions to CPT codes
    y_pred_labels = label_encoder.inverse_transform(y_pred)
    y_test_labels = label_encoder.inverse_transform(y_test)

    print(f"\n✅ {model_name} Results (Word2Vec):")
    print(classification_report(y_test_labels, y_pred_labels))

    # Confusion Matrix
    sns.heatmap(confusion_matrix(y_test_labels, y_pred_labels), annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name} (Word2Vec)')
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```

#LSTM

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional
# Tokenizer
tokenizer = Tokenizer(num_words=5000, oov_token="<OOV>")
tokenizer.fit_on_texts(df_notes['Description'])

# Text to Sequences
sequences = tokenizer.texts_to_sequences(df_notes['Description'])
padded_sequences = pad_sequences(sequences, maxlen=50, padding='post')

# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
labels = le.fit_transform(df_notes['Code'])

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, stratify=labels, random_state=42)



# LSTM Model
model = Sequential()
model.add(Embedding(input_dim=5000, output_dim=128, input_length=50))
model.add(Bidirectional(LSTM(64)))
model.add(Dense(len(le.classes_), activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Train
history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
# Predict
y_pred = model.predict(X_test)
y_pred_labels = np.argmax(y_pred, axis=1)

# Evaluate with fixed labels
print(classification_report(y_test, y_pred_labels, target_names=[str(c) for c in le.classes_]))

sns.heatmap(confusion_matrix(y_test, y_pred_labels), annot=True, fmt='d',
            xticklabels=[str(c) for c in le.classes_],
            yticklabels=[str(c) for c in le.classes_])
plt.title('Confusion Matrix - LSTM Model')
plt.show()

plot_training_curves(history)
```

```python
# Prepare input and target
X = df_notes['Description'].tolist()
y = df_notes['Code']

# Encode target labels
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split data
X_train, X_test, y_train_enc, y_test_enc = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

# Function to extract CLS embeddings from BERT
def get_bert_cls_embeddings(texts):
    embeddings = []
    for text in tqdm(texts, desc="Generating BERT embeddings"):
        inputs = tokenizer(text, return_tensors='pt', truncation=True, padding='max_length', max_length=128)
        with torch.no_grad():
            outputs = bert_model(**inputs)
        cls_embedding = outputs.last_hidden_state[:, 0, :].squeeze().numpy()
        embeddings.append(cls_embedding)
    return np.array(embeddings)


X_train_vec = get_bert_cls_embeddings(X_train)
X_test_vec = get_bert_cls_embeddings(X_test)
```

# Visualizations of Different Best Models with TF-IDF

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve, validation_curve
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Load and prepare data
X = df_notes['Description']
y = df_notes['Code']

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# TF-IDF vectorization
tfidf = TfidfVectorizer(max_features=3000)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# Initialize SVM model
svm_model = SVC(kernel='linear', probability=True, random_state=42)

# 1. Learning Curve Visualization
def plot_learning_curve(estimator, title, X, y, cv=5):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=-1,
        train_sizes=np.linspace(0.1, 1.0, 5),
        scoring='accuracy'
    )

    plt.figure(figsize=(10, 6))
    plt.title(title)
    plt.xlabel("Training examples")
    plt.ylabel("Accuracy Score")

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                    train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                    test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
            label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
            label="Cross-validation score")

    plt.legend(loc="best")
    plt.show()

# 2. Validation Curve for C parameter
def plot_validation_curve(estimator, title, X, y, param_name, param_range):
    train_scores, test_scores = validation_curve(
        estimator, X, y, param_name=param_name, param_range=param_range,
        cv=5, scoring="accuracy", n_jobs=-1
    )

    plt.figure(figsize=(10, 6))
    plt.title(title)
    plt.xlabel(param_name)
    plt.ylabel("Accuracy Score")

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()
    plt.semilogx(param_range, train_scores_mean, label="Training score", color="r")
    plt.fill_between(param_range, train_scores_mean - train_scores_std,
                    train_scores_mean + train_scores_std, alpha=0.2, color="r")
    plt.semilogx(param_range, test_scores_mean, label="Cross-validation score", color="g")
    plt.fill_between(param_range, test_scores_mean - test_scores_std,
                    test_scores_mean + test_scores_std, alpha=0.2, color="g")

    plt.legend(loc="best")
    plt.show()

# Train SVM and plot curves
print("Training SVM with TF-IDF...")
svm_model.fit(X_train_tfidf, y_train)

# Plot learning curve
plot_learning_curve(svm_model, "SVM Learning Curve (TF-IDF)", X_train_tfidf, y_train)

# Plot validation curve for C parameter
plot_validation_curve(svm_model, "SVM Validation Curve for C parameter (TF-IDF)",
                     X_train_tfidf, y_train, "C", np.logspace(-3, 3, 7))

# Evaluate model
y_pred = svm_model.predict(X_test_tfidf)
y_test_labels = label_encoder.inverse_transform(y_test)
y_pred_labels = label_encoder.inverse_transform(y_pred)

print("\nSVM Results (TF-IDF):")
print(classification_report(y_test_labels, y_pred_labels))

# Confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test_labels, y_pred_labels),
            annot=True, fmt='d', cmap='Blues')
plt.title('SVM Confusion Matrix (TF-IDF)')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## Random Forest with TF-IDF

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train and plot curves
print("Training Random Forest with TF-IDF...")
rf_model.fit(X_train_tfidf, y_train)

# Learning Curve
plot_learning_curve(rf_model, "Random Forest Learning Curve (TF-IDF)", X_train_tfidf, y_train)

# Validation Curve for n_estimators
plot_validation_curve(rf_model, "Random Forest Validation Curve for n_estimators (TF-IDF)",
                    X_train_tfidf, y_train, "n_estimators", [10, 50, 100, 200])

# Evaluate
y_pred = rf_model.predict(X_test_tfidf)
y_pred_labels = label_encoder.inverse_transform(y_pred)

print("\nRandom Forest Results (TF-IDF):")
print(classification_report(y_test_labels, y_pred_labels))

# Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test_labels, y_pred_labels),
            annot=True, fmt='d', cmap='Blues')
plt.title('Random Forest Confusion Matrix (TF-IDF)')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
